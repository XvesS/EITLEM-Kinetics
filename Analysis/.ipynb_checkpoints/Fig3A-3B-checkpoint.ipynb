# 画训练曲线

```python
import sys
sys.path.append('../')
from tools import getTensorBoardData
rawKcatTrainInfo = getTensorBoardData("../../IterTransferLearning/KCATModel/confirm-MACCS#openKKmInfer#openLog10-KCAT-train-1/logs/")
rawKmTrainInfo = getTensorBoardData("../../IterTransferLearning/KMModel/confirm-MACCS#openKKmInfer#openLog10-KM-train-1/logs/")
rawKKmTrainInfo = getTensorBoardData("../../KKM/TrainInfo/logs/initialTrain-MACCSKeys/")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
plt.figure(figsize=(4,4))
rc('font',**{'family':'Times New Roman',})
plt.rcParams['pdf.fonttype'] = 42
# plt.axes([0.12,0.12,0.83,0.83])
plt.tick_params(direction='in')
plt.tick_params(which='major',length=2)
plt.tick_params(which='major',width=1)
plt.plot(list(range(100))[::5],rawKcatTrainInfo['dev']['RMSE'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#159090', markersize=6,label='$K_{cat}$') 
plt.plot(list(range(100))[::5],rawKmTrainInfo['dev']['RMSE'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#b2182b', markersize=6,label='$K_{m}$')
plt.plot(list(range(100))[::5],rawKKmTrainInfo['dev']['RMSE'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#2166ac', markersize=6,label='$KK_{m}$')
plt.rcParams['font.family'] = 'Times New Roman'
plt.xticks(list(range(100))[::10])
# plt.yticks([0.5,0.7,0.9,1.1,1.3,1.5])
plt.xlabel('Epoch', fontsize=12, weight='bold')
plt.ylabel('Log10-scaled MAE', fontsize=12, weight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(frameon=False, prop={"size":12})
ax = plt.gca()
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)
ax.spines['top'].set_linewidth(0.5)
ax.spines['right'].set_linewidth(0.5)
plt.savefig("../Fig/Fig1-MACCSKeysTrainProceeRMSE.png", dpi=600,bbox_inches='tight')
print(rawKcatTrainInfo['dev']['RMSE'][-1])
print(rawKmTrainInfo['dev']['RMSE'][-1])
print(rawKKmTrainInfo['dev']['RMSE'][-1])
```

```python
plt.figure(figsize=(4,4))
rc('font',**{'family':'Times New Roman',})
plt.rcParams['pdf.fonttype'] = 42
# plt.axes([0.12,0.12,0.83,0.83])
plt.tick_params(direction='in')
plt.tick_params(which='major',length=2)
plt.tick_params(which='major',width=1)
plt.plot(list(range(100))[::5],rawKcatTrainInfo['dev']['R2'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#159090', markersize=6,label='$K_{cat}$') 
plt.plot(list(range(100))[::5],rawKmTrainInfo['dev']['R2'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#b2182b', markersize=6,label='$K_{m}$')
plt.plot(list(range(100))[::5],rawKKmTrainInfo['dev']['R2'][::5],linestyle='dashed',linewidth=1.75,marker='o',markerfacecolor='#2166ac', markersize=6,label='$KK_{m}$')
plt.rcParams['font.family'] = 'Times New Roman'
plt.xticks(list(range(100))[::10])
plt.ylim(0.2, 0.7)

# plt.yticks([0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, ])
plt.xlabel('Epoch', fontsize=12, weight='bold')
plt.ylabel('$R^{2}$', fontsize=12, weight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(frameon=False, prop={"size":12})
ax = plt.gca()
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)
ax.spines['top'].set_linewidth(0.5)
ax.spines['right'].set_linewidth(0.5)
plt.savefig("../Figure/Fig3B-MACCSKeysTrainProceeR2.png", dpi=600,  bbox_inches='tight')
print(rawKcatTrainInfo['dev']['R2'][-1])
print(rawKmTrainInfo['dev']['R2'][-1])
print(rawKKmTrainInfo['dev']['R2'][-1])
```
