```python
import torch
import sys
sys.path.append("../")
from DLkcatTools import dlkcatPredict
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from rdkit.Chem import AllChem
from rdkit import Chem
seqIndex = torch.load("../Data/KCAT/NewestFeature/index_seq")
smilesIndex = torch.load("../Data/KCAT/NewestFeature/index_smiles")
testPair = torch.load("../Data/KCATTestPairInfo")
# 定义函数
def shuffle_dataset(dataset, seed):
    np.random.seed(seed)
    np.random.shuffle(dataset)
    return dataset
def split_dataset(dataset, ratio):
    n = int(ratio * len(dataset))
    dataset_1, dataset_2 = dataset[:n], dataset[n:]
    return dataset_1, dataset_2
# 去除数据集中出现在DLkcat训练集的样本
# 加载数据
import json
with open("../../KCAT/DLKcat/Data/database/Kcat_combination_0918.json", 'r') as f:
    rawDlKcatData = json.load(f)
dlKcatData = []
for item in rawDlKcatData:
    if '.' not in item['Smiles'] and float(item['Value']) > 0:
        smiles = Chem.MolToSmiles(Chem.MolFromSmiles(item['Smiles']))
        seq = item['Sequence']
        dlKcatData.append(f'{seq}!{smiles}')
# 分割为训练集和测试集
dlKcatData = shuffle_dataset(dlKcatData, 1234)
dlTrainDataset, _ = split_dataset(dlKcatData, 0.8)
# dataset_dev, dataset_test = split_dataset(dataset_, 0.5)
# 去除测试集中存在dlkcat训练集中的样本
ignoreIndex = []
pairForDLkcat = []
for idx, item in enumerate(testPair):
    seq = seqIndex[item[0]]
    smiles = smilesIndex[item[1]]
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    queryItem = f'{seq}!{smiles}'
    if queryItem in dlTrainDataset:
        ignoreIndex.append(idx)
    else:
        pairForDLkcat.append(item)
```

```python
preValue, tarValue = dlkcatPredict(pairForDLkcat, seqIndex, smilesIndex)
r2_score(tarValue, preValue), np.sqrt(mean_squared_error(tarValue, preValue)), np.abs(tarValue - preValue).sum() / len(tarValue)
```

## KM预测测试

```python
"""
    1.从alex测试集中剔除我们数据集中纯在的数据样本
    2.匹配Brenda数据库中的Km值，检查是否相同, 剔除具有不相同的的样本
"""
import json
import pandas as pd
import numpy as np
import math
from rdkit import Chem
import json
from collections import defaultdict
testDf = pd.read_pickle("../../KM/KM_prediction/datasets/splits/test_data.pkl")
rawKmDf = pd.read_pickle("../../KM/KM_prediction/datasets/BRENDA_data/brenda_df_checkpoint_4.pkl")
testDf['Unirep'] = testDf['Unirep'].astype(str)
rawKmDf['Unirep'] = rawKmDf['Unirep'].astype(str)
mergeDf = pd.merge(testDf, rawKmDf, on=['KEGG ID', 'Unirep'], how='inner')
mergeDf = mergeDf.drop(columns=['log10_KM_y', 'substrate'])
mergeDf = mergeDf.drop_duplicates(subset=['KEGG ID', 'Sequence'])
mergeDf = mergeDf.drop(columns=["Unirep", "ECFP", "RDKit FP", "MACCS FP", "MW", "LogP", "GNN FP", "UNIPROT", "KEGG reaction ID"])
mergeDf = mergeDf.rename(columns={"log10_KM_x":'VALUE'})
# 转换
tempKmSet = []
for ind in mergeDf.index:
    kegg_id = mergeDf['KEGG ID'][ind]
    if not pd.isnull(kegg_id):
        try:
            mol = Chem.MolFromMolFile("../../KM/KM_prediction/datasets" +  "/mol-files/" + kegg_id + '.mol')
            if mol is not None:
                smiles = Chem.MolToSmiles(mol)
                seq = mergeDf['Sequence'][ind]
                k = f'{seq}_{smiles}'
                tempKmSet.append({
                        'ec':mergeDf['EC'][ind],
                        'organism':mergeDf['ORGANISM'][ind],
                        'substrate':mergeDf['SUBSTRATE'][ind],
                        'sequence':mergeDf['Sequence'][ind],
                        'smiles':Chem.MolToSmiles(mol),
                        'value':math.pow(10, mergeDf['VALUE'][ind]),
                        'type':'wild',
                        'source':'article'
                    })
        except:
            pass
# 去除掉数据集中冗余数据
tempValues = defaultdict(list)
for item in tempKmSet:
    seq = item['sequence']
    smiles = item['smiles']
    tempValues[f'{seq}_{smiles}'].append(item['value'])
saveKeys = []
def tolenrace(l, t):
    length = len(l)
    for i in range(length):
        for j in range(length):
            if abs(l[i] - l[j]) > t:
                return False
    return True
            
for k, v in tempValues.items():
    if tolenrace(v, 0.0001):
        saveKeys.append(k)
# 精简
curKey = set()
newKmSet = []
for item in tempKmSet:
    seq = item['sequence']
    smiles = item['smiles']
    k = f'{seq}_{smiles}'
    if k in saveKeys and k not in curKey:
        newKmSet.append(item)
        curKey.add(k)
with open("../../KM/Data/processed_km_value_data_deduplicate.json", 'r') as f:
    km_data = json.load(f)
km_df = pd.DataFrame(km_data)
new_km_df = pd.DataFrame(newKmSet)
df = pd.merge(km_df, new_km_df, how='inner', on=['sequence', 'smiles'])
df = df[np.abs(df['value_x'] - df['value_y'])<1e-2]
df = df.drop(columns=['ec_y', 'organism_y', 'substrate_y', 'value_y', 'type_y', 'source_x'])
df = df.rename(columns={'ec_x':'ec', 'organism_x':'organism', 'substrate_x':'substrate', 'value_x':'value', 'type_x':'type', 'source_y':'source'})
kmDataset = df.to_json(orient="records")
kmDataset = json.loads(kmDataset)
```

```python
'''剔除存在我们训练集中的样本'''
from rdkit import Chem
import sys
import torch
sys.path.extend(["../", "../../KCAT/scripts/"])
from predictor_iter import predict
from tools import getPairInfo, metric
from utils import get_pair_info, inference_distribute
# # 加载训练集序列和分子索引
ownIndexSeq = torch.load("../../../Kinetics/KM/Data/NewestFeature/index_seq")
ownIndexSmiles = torch.load("../../../Kinetics/KM/Data/NewestFeature/index_smiles")
train_pair_info = torch.load("../Data/KMTrainPairInfo")
train_pair_info, _, _ = inference_distribute(train_pair_info, [], [], None, True)
ownTrainPair = train_pair_info
# 加载训练集序列和分子索引
kmIndexSeq = torch.load("../../KM/Data/KM_feature/index_seq")
kmIndexSmiles = torch.load("../../KM/Data/KM_feature/index_smiles")
kmPairInfo = torch.load("../../KM/Data/KM_feature/test_info")
# 创建唯一值
trainInfoSet = set()
for item in ownTrainPair:
    seq = ownIndexSeq[item[0]]
    smiles = ownIndexSmiles[item[1]]
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    trainInfoSet.add(f"{seq}${smiles}")
testPairInfo = []
for item in kmPairInfo:
    seq = kmIndexSeq[item[0]]
    smiles = kmIndexSmiles[item[1]]
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    if f"{seq}${smiles}" not in trainInfoSet:
        testPairInfo.append(item)
# model_path = "../../../Kinetics/IterTransferLearning/KMModel/confirm-MACCS#openKKmInfer#openLog10-KM-train-8/model/pre-train-trainR2:0.9330-devR2=0.7161-RMSE=0.6963-MAE=0.4793-epoch=4"
model_path = "../../../Kinetics/IterTransferLearning/KMModel/confirm-MACCS#openKKmInfer#openLog10-KM-train-8/model/pre-train-trainR2:0.9330-devR2=0.7161-RMSE=0.6963-MAE=0.4793-epoch=4"
KineticsType = 'km'
Real, Pre = predict(
                KineticsType.upper(),
                model_path,
                testPairInfo,
                f"../../KM/Data/KM_feature/esm1v_t33_650M_UR90S_1_embeding_1280/",
                f"../../KM/Data/KM_feature/index_smiles",
                True,
                torch.device("cuda:1"),
                "MACCSKeys"
        )
metric(Pre, Real, True)
```

```python

import time
import os
import torch
from rdkit import Chem
import numpy as np
# # 加载训练集序列和分子索引
ownIndexSeq = torch.load("../../KM/Data/NewestFeature/index_seq")
ownIndexSmiles = torch.load("../../KM/Data/NewestFeature/index_smiles")
ownTrainPair = torch.load("../../paperAnalyse/Data/KMTrainPairInfo")

# 加载训练集序列和分子索引
kmIndexSeq = torch.load("../../KM/Data/KM_feature/index_seq")
kmIndexSmiles = torch.load("../../KM/Data/KM_feature/index_smiles")
kmPairInfo = torch.load("../../KM/Data/KM_feature/test_info")

# 创建唯一值
search = []
trainInfoSet = set()
for item in ownTrainPair:
    seq = ownIndexSeq[item[0]]
    smiles = ownIndexSmiles[item[1]]
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    trainInfoSet.add(f"{seq}${smiles}")
    search.append(seq)
query = []
testPairInfo = []
for item in kmPairInfo:
    seq = kmIndexSeq[item[0]]
    smiles = kmIndexSmiles[item[1]]
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    if f"{seq}${smiles}" not in trainInfoSet:
        testPairInfo.append(item)
        query.append(seq)

start_time = time.time()
torch.save(query, "./query")
torch.save(search, "./search")
!python ../getSimi.py -t 1 -q "./query" -s "./search"
results = torch.load("./simiResults")
similarity = np.array(list(results.values()))
print(f"0-40%:{(similarity<=0.4).sum() / len(similarity)}")
print(f"40-80%:{ ((similarity<=0.8) & (similarity>0.4)).sum() / len(similarity)}")
print(f"80-99%:{((similarity<=0.99) & (similarity>0.8)).sum() / len(similarity)}")
print(f"99-100%:{((similarity<=1) & (similarity>0.99)).sum() / len(similarity)}")
print(f"spend {time.time() - start_time} s")
```
